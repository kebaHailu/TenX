# Histograms and Bubble Charts 

### Histograms 

```python
# import library and data
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
 
benin = pd.read_csv('../../data/benin-malanville.csv')
sierraleone = pd.read_csv('../../data/sierraleone-bumbuna.csv')
togo = pd.read_csv('../../data/togo-dapaong_qc.csv')
```

#### Steps to follow 
- Create histogram for the specified variables 
- plot histograms for each variable in each dataset 
- plot for each city

```python
variables = ['GHI', 'DNI', 'DHI', 'WS', 'Tamb']


for variable in variables:
    plt.figure(figsize=(10, 6))
    
    # Plot for Benin
    plt.hist(benin[variable].dropna(), bins=30, alpha=0.5, label='Benin')
    
    # Plot for Sierra Leone
    plt.hist(sierraleone[variable].dropna(), bins=30, alpha=0.5, label='Sierra Leone')
    
    # Plot for Togo
    plt.hist(togo[variable].dropna(), bins=30, alpha=0.5, label='Togo')
    
    plt.title(f'Histogram of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.show()
```
